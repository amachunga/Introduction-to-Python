1. **(1 pt)** Create a function `diag_matrix(n)` which generate the following $n\times n$ array (you can use explicit loops):

**Example:**
```
>>> diag_matrix(3)
array([[ 0,  1,  2],
       [-1,  0,  1],
       [-2, -1,  0]])
>>> diag_matrix(4)
array([[ 0,  1,  2,  3],
       [-1,  0,  1,  2],
       [-2, -1,  0,  1],
       [-3, -2, -1,  0]])
>>> diag_matrix(6)
array([[ 0,  1,  2,  3,  4,  5],
       [-1,  0,  1,  2,  3,  4],
       [-2, -1,  0,  1,  2,  3],
       [-3, -2, -1,  0,  1,  2],
       [-4, -3, -2, -1,  0,  1],
       [-5, -4, -3, -2, -1,  0]])
```

2. **(1 pt)** Name `arr2` the array generated by `diag_matrix(8)`. Replace the values **strictly greater than 2** *OR* **strictly lower than -2** of `arr2` by 0.

````python
## WRITE IN THIS CELL YOUR CODE WITHOUT ANY ERRORS. YOU SHOULD GENERATE THE FUNCTION: diag_matrix
## question 1

def diag_matrix(n):
    x1=np.arange(n).reshape((1,n))
    xf=x1
    for i in range(1,n):
        xf=np.concatenate((xf,x1-i),axis=0)
    return xf
````

````python
## WRITE IN THIS CELL YOUR CODE WITHOUT ANY ERRORS. YOU SHOULD GENERATE THE LIST: arr2
## question 2

arr2=diag_matrix(8)
arr2[arr2>2]=0
arr2[arr2<-2]=0
````
